```python
from pylab import *
E=linspace(-0.25,0.25,501)
dE=E[1]-E[0]
kT=0.025
Ef=0
V=0
mu1=Ef+(V/2)
mu2=Ef-(V/2)

f1=1/(1+exp((E-mu1)/kT))
f2=1/(1+exp((E-mu2)/kT))
```

```python
plot(f1,E,'r',lw=2)
xlabel(' Fermi function $-->$ ',fontsize=15)
ylabel(' E - $\mu$ (eV) $-->$ ',fontsize=15)
xlim(-0.075,1.075)
ylim(-0.25,0.25)
tick_params(axis='both',which='major',labelsize=15,direction='in')
grid(color='b', alpha=0.5, ls='--',lw=0.5)
```
