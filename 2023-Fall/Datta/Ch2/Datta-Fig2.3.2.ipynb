### <font color = "navy" >2.3.1 Particle in a box </font>

```python
import numpy as np
import matplotlib.pyplot as plt

#Constants (all MKS, except energy which is in eV)
hbar=1.055e-34 
m=9.110e-31
epsil=8.854e-12
q=1.602e-19

#Lattice
Np=100
a=1e-10 
X=a*np.array(list(range(1,Np+1))) 
t0=(hbar**2)/(2*m*(a**2))/q

#analytical eigenvalues
L=a*(Np+1)
Ean=(((hbar*np.pi)**2)/(2*m*(L**2))/q)*np.array((list(range(1,Np+1))))*np.array((list(range(1,Np+1))));

#numerical eigenvalues
T=(2*t0*np.diag(np.ones((Np)))-(t0*np.diag(np.ones((Np-1)),1))-(t0*np.diag(np.ones((Np-1)),-1)))

[D,V]=np.linalg.eigh(T)
Enum=np.sort(D)

ind=np.array(list(range(0,100)))
#Probability
#alpha=1
E1=D[ind[0]]
psi1=abs(V[:,ind[0]])
P1=psi1*np.conj(psi1)

E2=D[ind[24]]
psi2=abs(V[:,ind[24]])
P2=psi2*np.conj(psi2)
```

```python
fig=plt.figure(figsize=(16,5))

ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

#for figure 2.3.2(a)
ax1.plot(Ean,'b',lw=2,label='Analytical') #Part(a)
ax1.plot(Enum,'bx',lw=1,label='Numerical') #Part(a)
ax1.set_xlabel(' Eigenvalue Number, ' r'$\alpha$',fontsize=14) #Part(a)
ax1.set_ylabel(' E (eV) ',fontsize=14) #Part(a)
ax1.set_xlim(0,100)
ax1.set_ylim(0,40)
ax1.tick_params(axis='both',which='major',direction='in')
#ax1.text(45,20,'Analytical',fontsize=14)
#ax1.text(65,7,'Numerical',fontsize=14)
ax1.grid(alpha=0.5, ls='--')
ax1.legend(loc=2,fontsize=14)

#for figure 2.3.2(b)
ax2.plot(P1,'r',lw=3,label=r'$\alpha$=1') #Part(b)
ax2.plot(P2,'r',lw=1,label=r'$\alpha$=25') #Part(b)
ax2.set_xlabel(' Lattice site number ',fontsize=14) # Part(b)
ax2.set_ylabel(' Probability ',fontsize=14) # Part(b)
ax2.set_xlim(0,100)
ax2.set_ylim(0,0.02)
ax2.tick_params(axis='both',which='major',direction='in')
#ax2.annotate(r'$\alpha$=1', xy=(40,0.019), xytext=(30,0.022),
#                 arrowprops=dict(facecolor='black',shrink=0.03),fontsize=14)
#ax2.annotate(r'$\alpha$=25', xy=(78,0.019), xytext=(85,0.022),
#                 arrowprops=dict(facecolor='black',shrink=0.03),fontsize=14)
ax2.grid(alpha=0.5, ls='--')
ax2.legend(loc=2,fontsize=14)
```
