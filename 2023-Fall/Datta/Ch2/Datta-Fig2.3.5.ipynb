### <font color = "navy" >2.3.1 Particle in a box </font>

#### Boundary conditions

```python
import numpy as np
import matplotlib.pyplot as plt

#Constants (all MKS, except energy which is in eV)
hbar=1.055e-34
m=9.110e-31
epsil=8.854e-12
q=1.602e-19;

#Lattice
Np=100
a=1e-10
X=a*np.array(list(range(1,Np+1)))

t0=(hbar**2)/(2*m*(a**2))/q

#numerical eigenvalues
T=(2*t0*np.diag(np.ones((Np)))-(t0*np.diag(np.ones((Np-1)),1))-(t0*np.diag(np.ones((Np-1)),-1)))
T[0,Np-1]=-t0
#T[Np-1,0]=-t0

[D,V]=np.linalg.eigh(T)
Enum=np.sort(D)

t1=(hbar**2)/(2*m*(a**2))/q

P=(2*t1*np.diag(np.ones((Np)))-(t1*np.diag(np.ones((Np-1)),1))-(t1*np.diag(np.ones((Np-1)),-1)))
P[Np-1,0]=-t1

[D,V]=np.linalg.eigh(P)
Enum2=np.sort(D)

ind=np.array(list(range(0,100)))
```

```python
#for figure 2.3.5
plt.figure()

plt.plot(Enum,'bx',lw=2)
plt.plot(Enum2,'rx',lw=2)
plt.xlabel(' Eigenvalue Number, 'r'$\alpha$ ',fontsize=14)
plt.ylabel(' E (eV)',fontsize=14)
plt.xlim(0,100)
plt.ylim(-2,16)
#plt.xlim(40,60)
#plt.ylim(4,10)
plt.tick_params(axis='both',which='major',labelsize=15,direction='in')
plt.grid(alpha=0.5, ls='--')
```
