# 3. Self-consistent field 

## 3.3 Bonding

```python
import numpy as np
import matplotlib.pyplot as plt

#Constants (all MKS, except energy which is in eV)
hbar=1.055e-34 
m=9.110e-31
epsil=8.854e-12
q=1.602e-19

a0=4*np.pi*epsil*hbar*hbar/(m*q*q)
E0=q/(8*np.pi*epsil*a0);

R0=0.05*np.array(list(range(1,201)))
a=(-2*E0)*(1-(np.exp(-2*R0)*(1+R0)))/R0
b=(-2*E0)*np.exp(-R0)*(1+R0)
s=(1+R0+((R0**2)/3))*np.exp(-R0)
Uee=(2*E0)/np.sqrt(1+(R0**2))
UNN=(2*E0)/R0

EB0=(a+b)/(1+s)
R=a0*R0
```

```python
#for figure 3.3.4
plt.figure()

plt.plot(R,EB0,'r--',lw=1,label='$E_{B0}-E_0$')
plt.plot(R,Uee,'rx',lw=1,label='$U_{e,e\'}$')
plt.plot(R,UNN,'r',lw=1,label='$U_{N,N\'}$')
plt.plot(R,(2*EB0)+UNN+Uee,'b+',lw=1,label='Binding Energy')
plt.xlabel(' R (m) ',fontsize=14)
plt.ylabel(' Energy (eV) ',fontsize=14)
plt.xlim(0,4e-10)
plt.ylim(-25,25)
plt.tick_params(axis='both',which='major',labelsize=15,direction='in')
plt.grid(alpha=0.5, ls='--')
plt.legend(loc=1,fontsize=12)

#plt.text(0.8e-10,20,'$U_{N,N\'}$',fontsize=12)
#plt.text(0.63e-10,10,'$U_{e,e\'}$',fontsize=12)
#plt.text(0.3e-10,-8.5,'Binding Energy',fontsize=12)
#plt.text(3.1e-10,-9,'$E_{B0}-E_0$',fontsize=12)
```
