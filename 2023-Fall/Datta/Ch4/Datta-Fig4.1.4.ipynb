```python
import numpy as np
import matplotlib.pyplot as plt

#Constants (all MKS, except energy which is in eV)
hbar=1.055e-34; m=9.110e-31; epsil=8.854e-12; q=1.602e-19;
a0=4*np.pi*epsil*hbar*hbar/(m*q*q); E0=q/(8*np.pi*epsil*a0);

#Basis
L=0.074e-9/a0; s=np.exp(-L)*(1+L+((L**2)/3));
r=np.linspace(-2e-10,+2e-10,101); r0=r/a0;
psi=np.sqrt(1/(np.pi*(a0**3)))*(np.exp(-abs(r0-(L/2)))+np.exp(-abs(r0+(L/2))));
n=2*psi*np.conj(psi)/(2*(1+s));

a=-2*E0*(1-((1+L)*np.exp(-2*L)))/L;
b=-2*E0*(1+L)*np.exp(-L);
EB0=-E0+((a+b)/(1+s));
```

```python
plt.plot(r,n,'r',lw=2)
plt.xlabel(' $Z$ (m) ',fontsize=15)
plt.ylabel(' Electron density ('r'$\ /m^3$'') ',fontsize=14)
plt.xlim(-2e-10,2e-10)
plt.ylim(0,2e30)
plt.tick_params(axis='both',which='major',labelsize=14,direction='in')
plt.grid( alpha=0.5, ls='--')
```
