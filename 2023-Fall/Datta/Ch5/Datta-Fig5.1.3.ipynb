```python
import numpy as np
import matplotlib.pyplot as plt

k=np.linspace(-1,1,21)
E0=0
a=-1  #Ess
b=1   #Ess'
E1=E0+(2*a*np.cos(np.pi*k))
E2=E0+(2*b*np.cos(np.pi*k))
```

```python
#for figure 5.1.3
plt.figure()

plt.plot(k,E1,'r-o',lw=1)
plt.plot(k,E2,'b-x',lw=1)
plt.xlabel('ka/$\pi$ ',fontsize=14)
plt.ylabel(' Energy (eV) ',fontsize=14)
plt.xlim(-1,1)
plt.ylim(-2,2)
plt.tick_params(axis='both',which='major',labelsize=14,direction='in')
plt.grid(alpha=0.5, ls='--')
```
