```python
import numpy as np
import matplotlib.pyplot as plt

k=np.linspace(-1,1,101)
E0=0
a=2  #Ess
b=1  #Ess'
E1=E0+np.sqrt((a**2)+(b**2)+(2*a*b*np.cos(np.pi*k)))
```

```python
#for figure 5.1.5
plt.figure()

plt.plot(k,E1,'r',lw=1)
plt.plot(k,-E1,'r',lw=2)
plt.xlabel(' k (in units of $\pi$/a) ',fontsize=14)
plt.ylabel(' Energy (eV) ',fontsize=14)
plt.xlim(-1,1)
plt.tick_params(axis='both',which='major',labelsize=14,direction='in')
plt.grid(alpha=0.5, ls='--')
```
