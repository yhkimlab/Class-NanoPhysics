## <font color = "navy" >10.4 Supplementary notes: phonons </font> <a id="10.4"></a>

```python
%reset -f
%matplotlib inline
from pylab import *
```

```python
beta=pi*linspace(-1,1,201); w0=1;
y=sqrt(2*w0*(1-cos(beta)));
```

```python
plot(beta,y,'r',lw=2)
xlabel(r' $\beta a  -->$ ',fontsize=15)
ylabel(' $\omega /\omega_0   -->$ ',fontsize=15)
xlim(-4,4)
ylim(0,2)
tick_params(axis='both',which='major',labelsize=15,direction='in')
grid(color='b', alpha=0.5, ls='--', lw=0.5)
```

<font color = "stone"><font size = "4" > __Fig. 10.4.4__ (b) Dispersion law $\omega /\omega_0$ vs. $\beta a$. </font>

```python
%reset -f
%matplotlib inline
from pylab import *
import cmath
```

```python
beta=pi*linspace(-1,1,201); w1=1; w2=2;

#Initializing arrays
D1=zeros((201)); D2=zeros((201)); 

for n in range(0,201):
    A=[[w1+w2, w1+(w2*exp(-1j*beta[n]))], [w1+(w2*exp(1j*beta[n])), w1+w2]];
    [D,V]=eig(A); D=sort(real(D));
    D1[n]=real(sqrt(D[0])); D2[n]=real(sqrt(D[2-1]));
```

```python
plot(beta,D1,'r',lw=2)
plot(beta,D2,'r',lw=2)
xlabel(r' $\beta a -->$ ',fontsize=15)
ylabel(' $\omega /\omega_1 -->$ ',fontsize=15)
xlim(-4,4)
ylim(0,2.5)
tick_params(axis='both',which='major',labelsize=15,direction='in')
text(-0.7,1.9,'Optical \n phonon',fontsize=15)
text(-0.7,0.6,'Acoustic \n phonon',fontsize=15)
grid(color='b', alpha=0.5, ls='--', lw=0.5)
```

<font color = "stone"><font size = "4" > __Fig. 10.4.5__ (b) Dispersion law $\omega /\omega_1$ vs. $\beta a$ showing acoustic and optical phonons assuming $\omega_2$ = $\sqrt{2}\omega_1$. </font>
